```python
import os
from PIL import Image
import pytesseract
import cv2
import matplotlib as mpl
mpl.use('TkAgg')
from matplotlib import pyplot as plt
from skimage.filters import (threshold_otsu, threshold_niblack,
                             threshold_sauvola)
from skimage.restoration import (denoise_tv_chambolle, denoise_bilateral,
                                 denoise_wavelet, estimate_sigma)
from skimage import io
from skimage.color import rgb2gray
from skimage import img_as_uint
from skimage.exposure import equalize_adapthist
```

```python
%matplotlib inline
```

```python
path_image = '../01_alb_id.tif'
```

```python
image = rgb2gray(io.imread(path_image))
```

```python
plt.imshow(image, cmap=plt.cm.gray)
```

```python
image_CLAHE = equalize_adapthist(image)
plt.imshow(image_CLAHE, cmap=plt.cm.gray)
```

```python
denoised = denoise_wavelet(image)
plt.imshow(denoised, cmap=plt.cm.gray)
```

```python
binary_global = denoised > threshold_otsu(denoised)

thresh_niblack = threshold_niblack(denoised, window_size=25, k=0.8)
thresh_sauvola = threshold_sauvola(denoised, window_size=25)

binary_niblack = denoised > thresh_niblack
binary_sauvola = denoised > thresh_sauvola
```

```python
plt.figure(figsize=(16, 14))
plt.subplot(2, 2, 1)
plt.imshow(denoised, cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Global Threshold')
plt.imshow(binary_global, cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(binary_niblack, cmap=plt.cm.gray)
plt.title('Niblack Threshold')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(binary_sauvola, cmap=plt.cm.gray)
plt.title('Sauvola Threshold')
plt.axis('off')

plt.show()
```

```python
io.imsave('../preprocessed_images/global_noise.png', img_as_uint(binary_global))
io.imsave('../preprocessed_images/niblack_noise.png', img_as_uint(binary_niblack))
io.imsave('../preprocessed_images/sauvola_noise.png', img_as_uint(binary_sauvola))
```
